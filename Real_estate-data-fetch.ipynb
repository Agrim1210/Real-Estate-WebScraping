```python
from bs4 import BeautifulSoup # Import for Beautiful Soup
import requests # Import for requests
import lxml # Import for lxml parser
import pandas as pd # Import for Pandas
import matplotlib.pyplot as plt # Import for Matplotlib Pyplot
from urllib.request import urlopen
import numpy as np
```

```python
main_link = "https://www.magicbricks.com/low-budget-flats-for-sale-in-new-delhi-pppfs"

headers={'User-Agent':'Mozilla/5.0 (platform; rv:geckoversion) Gecko/geckotrail Firefox/firefoxversion'}
```

```python
# Request for the URL
page = requests.get(main_link,headers=headers)
# We can also check the response code
print(page.status_code)
# Make it a soup
soup = BeautifulSoup(page.text,"lxml")
# Display Soup ( Main Page)
# print(soup.prettify)
```

```python
property_info=soup.find_all('div',class_='mb-srp__card__info')
price_info=soup.find_all('div',class_='mb-srp__card__price')
```

```python
property_type=[]
area=[]
moving_status=[]
floor=[]
transaction_type=[]
furnishing=[]
direction_facing=[]
description=[]
price=[]
price_per_sq_ft=[]
for i in property_info:
    property_type.append(i.find_all('h2',class_='mb-srp__card--title')[0].text[:6])
    area.append(i.find_all('div',class_='mb-srp__card__summary--value')[0].text)
    moving_status.append(i.find_all('div',class_='mb-srp__card__summary--value')[1].text)
    floor.append(i.find_all('div',class_='mb-srp__card__summary--value')[2].text)
    transaction_type.append(i.find_all('div',class_='mb-srp__card__summary--value')[3].text)
    furnishing.append(i.find_all('div',class_='mb-srp__card__summary--value')[4].text)
    direction_facing.append(i.find_all('div',class_='mb-srp__card__summary--value')[5].text)
    description.append(i.find_all('div',class_='mb-srp__card--desc--text')[0].text.strip())

for i in price_info:
    price.append(i.find_all('div',class_='mb-srp__card__price--amount')[0].text)
    price_per_sq_ft.append(i.find_all('div',class_='mb-srp__card__price--size')[0].text)

    
    

    
    

```

```python
d = {'Property_Type': property_type,
     'Area':area,
     'Status':moving_status,
     'Floor':floor,
     'Transaction_Type':transaction_type,
     'Property_Furnishing':furnishing,
     'Direction_Facing':direction_facing,
     'Description':description,
     'Price':price,
     'Price_per_sqft':price_per_sq_ft}

#For page 1 only
df=pd.DataFrame(d)
df.head()
```
